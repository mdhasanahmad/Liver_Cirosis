```python
# Split into features and target
X = df.drop('Category', axis=1)
y = df['Category']
```

```python
# Debugging information
print(f"Shape of X: {X.shape}")
print(f"Shape of y: {y.shape}")
```

```python
# Define the number of features to select
N = 10
```

```python
# Feature selection methods
def pca_selection(X, n_features):
    pca = PCA(n_components=n_features)
    pca.fit(X)
    return np.argsort(pca.explained_variance_ratio_)[-n_features:]

def rfe_selection(X, y, n_features):
    rfe = RFE(estimator=LinearRegression(), n_features_to_select=n_features)
    rfe.fit(X, y)
    return np.where(rfe.support_)[0]
def pearson_correlation(X, y, n_features):
    corrs = [pearsonr(X.iloc[:, i], y)[0] for i in range(X.shape[1])]
    return np.argsort(np.abs(corrs))[-n_features:]


def ridge_selection(X, y, n_features):
    ridge = Ridge(alpha=1.0)
    ridge.fit(X, y)
    importances = np.abs(ridge.coef_)
    return np.argsort(importances)[-n_features:]

def variance_threshold_selection(X, threshold):
    selector = VarianceThreshold(threshold)
    selector.fit(X)
    return np.where(selector.variances_ > threshold)[0]
```

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.feature_selection import RFE, VarianceThreshold
from sklearn.linear_model import Ridge, LinearRegression
from scipy.stats import pearsonr
```

```python


# Apply feature selection methods
selected_features = {
    'PCA': pca_selection(X, N),
    'RFE': rfe_selection(X, y, N),
    'Pearson': pearson_correlation(X, y, N),
    'Ridge': ridge_selection(X, y, N),
    'VarianceThreshold': variance_threshold_selection(X, 0.1)
}
```

```python
# Aggregate and count feature selections
feature_counts = np.zeros(X.shape[1])
for method, features in selected_features.items():
    print(f"Selected features by {method}: {features}")
    feature_counts[features] += 1

# Define ensemble methods
def ensemble_selection(counts, threshold):
    return np.where(counts > threshold)[0]
```
